```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 
sns.set(color_codes=True)
```

```python
df=pd.read_csv("StudentsPerformance.csv")
df
```

```python
by_gender = df.groupby('gender')
by_gender.size().plot(kind='bar')
```

```python
by_race_ethnicity= df.groupby('race/ethnicity')
by_race_ethnicity.size().plot(kind='bar')
```

```python
by_parental_level_of_education = df.groupby('parental level of education')
by_parental_level_of_education.size().plot(kind='bar')
```

```python
by_lunch = df.groupby('lunch')
by_lunch.size().plot(kind='bar')
```

```python
by_test_preparation_course= df.groupby('test preparation course')
by_test_preparation_course.size().plot(kind='bar')
```

```python
df.describe()
```

```python
df.mean().plot(kind="bar")
```

```python
df.hist()
```

```python
plt.xlabel("Maths Score")
plt.ylabel("Student's Count")
maths_score_male =list(df[df['gender']=='male']['math score'])
maths_score_female =list(df[df['gender']=='female']['math score'])
plt.hist([maths_score_male,maths_score_female],  rwidth=0.95, color=['green','orange'],label=['Male','Female'])
plt.legend()
```

```python
plt.xlabel("Reading Score")
plt.ylabel("Student's Count")
reading_score_male =list(df[df['gender']=='male']['reading score'])
reading_score_female =list(df[df['gender']=='female']['reading score'])
plt.hist([reading_score_male,reading_score_female],  rwidth=0.95, color=['green','orange'],label=['Male','Female'])
plt.legend()
```

```python
plt.xlabel("Writing Score")
plt.ylabel("Student's Count")
writing_score_male =list(df[df['gender']=='male']['writing score'])
writing_score_female =list(df[df['gender']=='female']['writing score'])
plt.hist([writing_score_male,writing_score_female],  rwidth=0.95, color=['green','orange'],label=['Male','Female'])
plt.legend()
```

```python
by_m_gender = df.groupby('gender')
by_m_gender.mean().plot(kind='bar')

by_m_test_preparation_course= df.groupby('test preparation course')
by_m_test_preparation_course.mean().plot(kind='bar')

by_m_race_ethnicity= df.groupby('race/ethnicity')
by_m_race_ethnicity.mean().plot(kind='bar')

by_m_lunch = df.groupby('lunch')
by_m_lunch.mean().plot(kind='bar')

by_m_parental_level_of_education = df.groupby('parental level of education')
by_m_parental_level_of_education.mean().plot(kind='bar')
```
